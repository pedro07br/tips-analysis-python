```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print(tips.head(),"\n")
print(tips.info(),"\n")
print(tips.describe(),"\n")
```

```python
print("Media do total_bill")
media = tips["total_bill"].mean()
print(round(media,2))

print("\nNumero maximo do total_bill")
max = tips["total_bill"].max()
print(max)

print("\nMediana do tip")
mediana = tips["tip"].std()
print(round(mediana,2))

print("\nQuantos registros correspondem a cada dia da semana.")
dias = tips["day"].value_counts()
print(dias)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [1, 1]})

sns.histplot(tips["tip"], bins=15, ax=axes[0], kde=False, color="red")

sns.boxplot(x="sex", y="total_bill", data=tips, ax=axes[1], palette="pastel")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker", size="size", sizes=(20,200))
plt.show()
```

```python
media = tips.groupby("day")["tip"].mean()
print(media)

plt.figure(figsize=(8,6))
sns.barplot(x=media.index, y=media.values)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=tips["total_bill"])
plt.show()


contas_altas = tips[tips["total_bill"] > 50]
print(contas_altas)
```

```python
colunas = ["total_bill", "tip", "size"]

matriz_corr = tips[colunas].corr()
print(matriz_corr)

plt.figure(figsize=(6,5))
sns.heatmap(matriz_corr, annot=True)
plt.show()
```

```python
media_tip = tips.groupby(["sex", "smoker"])["tip"].mean()
print("Mdia de gorjetas por sexo e fumante:")
print(media_tip)

media_tip_reset = media_tip.reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x="sex", y="tip", hue="smoker", data=media_tip_reset)
plt.show()
```

```python
fumantes = tips[tips["smoker"] == "Yes"]
nao_fumantes = tips[tips["smoker"] == "No"]

plt.figure(figsize=(8,6))
sns.histplot(fumantes["total_bill"], bins=15, color="red", label="Fumantes", kde=False, alpha=0.5)
sns.histplot(nao_fumantes["total_bill"], bins=15, color="blue", label="No Fumantes", kde=False, alpha=0.5)

plt.legend()
plt.show()
```

```python
tips["percent_tip"] = tips["tip"] / tips["total_bill"] * 100

plt.figure(figsize=(8,6))
sns.scatterplot(x="size", y="percent_tip", data=tips)
plt.show()
```
